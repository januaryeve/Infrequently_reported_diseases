This notebook exists so I can play around with pulling data and ploting stuff

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('infrequently_reported_notifiable_diseases_data.csv')
df.head()
```

```python
plt.scatter(x=df['MMWR week'], y=df['Total cases reported for pervious years  2017'])
plt.show()
```

```python
import numpy as np
df.plot.hist(x= "Disease", y= "Total cases reported for pervious years  2017", color = 'magenta')

plt.ylabel("Total Cases Reported")
plt.show()
```

In this case, the histogram is a better representation of the data.

Now lets play with color...

```python
df.plot.hist(x= "Disease", y="Total cases reported for pervious years  2017", color='green')
```

```python
df.groupby(['Disease'], as_index = False).sum().plot(kind = 'bar', x = 'Disease', y = 'Total cases reported for pervious years  2017', color = 'purple')
plt.show()
```
